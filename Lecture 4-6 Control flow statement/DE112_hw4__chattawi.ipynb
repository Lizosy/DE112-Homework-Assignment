Write code to compute polygon area according to the type of polygon (Triangle, Square, Rectangle, or Kite)
* Take input string named polygon_type from a user
* check if polygon_type is valid
  * if yes, take number of required inputs from a user and compute area of that polygon according to the given formula
  * if polygon_type is not valid , print a message indicating that it's not a valid polygon name

Example

```
Polygon type: Triangle
Base: 2
Height: 10
Expected area of triangle: 10
```


```
Polygon type: Rectangle
Width: 20
Length: 10
Expected area of rectangle: 200
```

```
Polygon type: xyz
This is not one of the valid polygon name!
(The valid polygon name must be either Triangle, Square, Rectangle, or Kite)
```

````python
user_input = input("Enter the Polygon type :")
polyhon_tybe = ["Triangle", "Square","Rectangle","Kite"]

if user_input in polyhon_tybe:
    match user_input:
        case "Triangle":
            Base = input("Base:")
            Height = input("Height :")
            result = 1/2*int(Base)*int(Height)
            print("Expected area of rectangle", result)
        case "Square":
            Side = input("side:")
            result = int(Side)**2
            print("Expected area of Square", result)
        case "Rectangel":
            length = input("length:")
            width = input("width:")
            result = int(length)*int("width")
            print("Expected area of Rectangel:", result)
        case "kite" :
            d1 = input("d1:")
            d2 = input("d2:")
            result = int(d1)*int(d2)/2
            print("Expected area of kite :", result)
else:
     print("indicating that it's not a valid polygon name")
````

#3.
Write code to take calendar year from user and determine whether that is a leap year or not.

***To be a leap year, the year number must be divisible by four – except for end-of-century years, which must be divisible by 400. This means that the year 2000 was a leap year, although 1900 was not. 2020, 2024 and 2028 are all leap years.

**Example**
```
Calendar year: 1800
Expected output: Not a leap year
```
```
Calendar year: 1900
Expected output: Not a leap year
```
```
Calendar year: 2000
Expected output: A leap year
```
```
Calendar year: 2028
Expected output: A leap year
```


````python
user_input = input("Calendar year")

if (int(user_input)% 400 == 0) and (int(user_input) % 100 == 0):
    print("A leap year")

elif (int(user_input) % 4 ==0) and (int(user_input) % 100 != 0):
    print("A leap year")
else:
  print("Not a leap year")
````

````python
import numpy as np

user_input = input("Spending record: ").split()
my_array = np.array(user_input, dtype=int)
modified_list = list(my_array)
listcount = 0

for num in modified_list:
    if num >= 1000:
        listcount += 1

print("Eligible transactions:", listcount)

while listcount < 3:
    user_input = input("Spending record: ")
    transaction = int(user_input)
    if transaction >= 1000:
        listcount += 1

if listcount >= 3:
    print("Congratulations! You have received a 200 baht refund to your account.")
````
